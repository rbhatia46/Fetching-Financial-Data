### Getting data for multiple tickers

```python
tickers = ["AMZN","TSLA","MSFT","GOOG","HDFCBANK.NS","RELIANCE.NS"]
start_date = datetime.datetime.today() - datetime.timedelta(360)
end_date = datetime.datetime.today()
```

```python
close_price = pd.DataFrame()
for ticker in tickers:
    close_price[ticker] = yf.download(ticker,start_date,end_date)['Adj Close']
```

```python
close_price.head()
```
